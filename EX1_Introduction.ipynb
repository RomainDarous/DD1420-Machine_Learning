```python
def print_matrix(X):
  """
  prints the elements of a matrix
  input: X (n x m array)
  output: prints elements to the screen
  """
  for row in X:
    for val in row:
      print('{:4}'.format(val), end='')
    print('')
```

```python
def matrix_transpose(X):
  """
  Returns the transpose of a 2D matrix
  input: X (n x m array)
  output: result (m x n transposed array)
  """

  (n, m) = np.shape(X)
  result = np.zeros((m, n), dtype = X.dtype)

  for i in range(n) :
    for j in range (m) :
      result[i,j] = X[j, i]
        
  return result
```

```python
def matrix_multiply(A, B) :
    """
    Returns the multiplication of a 2D matrix
    input: A (n x m array), B (m x p array), 
    output: result (m x p array, the multiplication of the two matrix)
    """

    (n, m1) = A.shape
    (n2, p) = B.shape

    assert (m1 == n2), "Error, matrice shapes do not match !"

    result = np.zeros((n, p), dtype = A.dtype)

    for i in range (n) :
        for j in range(p) :
            result[i,j] = sum(A[i,k]*B[k,j] for k in range(m1))

    return result

C2 = matrix_multiply(matrix_transpose(A), B)
print('C2 = ')
print_matrix(C2)
assert (C1 == C2).all(), "Error, matrices should match!"
```
